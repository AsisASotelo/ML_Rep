```python
def load_dataset(filename):
    DIR_PATH = '../Datasets/' + filename
    # Load data with relative path
    data = pd.read_csv(DIR_PATH)
    
    # Convert to numpy array
    
    data_np = data.values
    
    #Retreive numpy array from the Dataframe
    X = data_np[:,:-1]
    y = data_np[:,-1]
    
    #Format all feature types as strings
    
    X = X.astype(str)
    
    #Reshape Target
    y = y.reshape((len(y),1))
    
    return X,y
    
```

```python
from sklearn.model_selection import train_test_split

# Load Data

X,y = load_dataset("breast-cancer.csv")



# Train/Test Split 

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.33,random_state=1)
```
